**Data Pre-Processing**

```python
# Lets perform the data cleaning
df_House_Price1.columns
```

```python
dictionary={
    'price':'Price',
    'resid_area':'Residential Area',
    'air_quality':'Air Quality',
    'room_number':'Room Number',
    'age':'Age',
    'distance':'Distance',
    'teachers':'Teachers',
    'poor_prop':'Poor_Property',
    'airport':'Airport',
    'n_hos_beds':'N_Hospital_Beds',
    'n_hot_rooms':'N_Hotet_Rooms',
    'waterbody':'Waterbody',
    'rainfall':'Rainfall',
    'bus_ter':'Bus_Ter',
    'sold':'Sold'
}
```

```python
# perform the rename action
df_House_Price1.rename(columns=dictionary)
```

```python
# Do the authentic check
df_House_Price1.columns
```

**Null Value Handling**

```python
# Find the number of missing records
df_House_Price1.isnull().sum()
```

```python
# Find the percentage
df_House_Price1.isnull().sum()/df_House_Price1.shape[0]*100
```

```python
# find the mode of waterbody column
df_House_Price1.waterbody.mode()
```

```python
# Lets treat the null values in the waterbody column
df_House_Price1.waterbody=df_House_Price1.waterbody.fillna('River')
```

```python
# Do the authentic check
df_House_Price1.isnull().sum()
```

```python
# In N_Hospital_Beds, we have less than 10% of data is missing
df_House_Price1.dropna(inplace=True)
```

```python
df_target=df_House_Price1['price']
```

```python
df_features=df_House_Price1.drop('price',axis=1)
```

```python
# Segregate the data based on the datatype
df_num=df_House_Price1.select_dtypes(include='number')
df_cat=df_House_Price1.select_dtypes(include='object')
```

```python
# show the numerical data
df_num.head()
```

```python
df_cat.head()
```

**Perform Encoding on categorical variables**

```python
# Use Dummy Encoding
df_encoded=pd.get_dummies(df_cat)
# Reset the index
df_encoded=df_encoded.reset_index()
```

```python
# show the encoded data
df_encoded.head()
```

**Perform Scalling On Numerical Data**

```python
# import the standard scaler
from sklearn.preprocessing import StandardScaler

# Initiate the standard scaler
ss=StandardScaler()

# scale the value present in numerical dataframe
scaled=ss.fit_transform(df_num)

# Numerate the data into the dataframe
scaled_df=pd.DataFrame(scaled,columns=df_num.columns)

# Reset the index
scaled_df=scaled_df.reset_index()
```

```python
# show the data
scaled_df.head()
```

**Perform the Concatenate Operation**

```python
# Concate encoded variable and scaled variable
df_independent=pd.concat([df_encoded,scaled_df],axis=1)
```

```python
# show the data
df_independent.head()
```

**Split the Data for Training and Testing**

```python
# import the train test split
from sklearn.model_selection import train_test_split
```

```python
# split the data
x_train,x_test,y_train,y_test=train_test_split(df_independent,
                                               df_target,
                                               test_size=0.2)
```

```python
# check the size
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

**Apply Linear Regression**

```python
# import the model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# initiate the model
LR=LinearRegression()
```

```python
# Train the linear regression model
model_lr=LR.fit(x_train,y_train)
```

```python
# perform the prediction
predicted=model_lr.predict(x_test)
```

```python
# show the predicted value
predicted
```

```python
# Evaluate the model
round(mean_squared_error(predicted,y_test),3)
```
