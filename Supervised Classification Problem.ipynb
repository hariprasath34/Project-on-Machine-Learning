```python
# import numerical python
import numpy as np

# import the pandas
import pandas as pd

#import the seaborn
import seaborn as sns

#import the matplotlib
import matplotlib.pyplot as plt
```

```python
# load the data
df_House_Price1 = pd.read_csv('House_Price1.csv')
```

```python
# show the first five observations of the data
df_House_Price1.head()
```

```python
dictionary={
    'price':'Price',
    'resid_area':'Residential Area',
    'air_quality':'Air Quality',
    'room_number':'Room Number',
    'age':'Age',
    'distance':'Distance',
    'teachers':'Teachers',
    'poor_prop':'Poor_Property',
    'airport':'Airport',
    'n_hos_beds':'N_Hospital_Beds',
    'n_hot_rooms':'N_Hotet_Rooms',
    'waterbody':'Waterbody',
    'rainfall':'Rainfall',
    'bus_ter':'Bus_Ter',
    'sold':'Sold'
}
```

```python
# perform the rename action
df_House_Price1.rename(columns=dictionary)
```

```python
# Lets treat the null values in the waterbody column
df_House_Price1.waterbody=df_House_Price1.waterbody.fillna('River')
```

```python
# In N_Hospital_Beds, we have less than 10% of data is missing
df_House_Price1.dropna(inplace=True)
```

```python
df_target=df_House_Price1['Sold']
```

```python
df_features=df_House_Price1.drop('Sold',axis=1)
```

```python
# Segregate the data based on the datatype
df_num=df_House_Price1.select_dtypes(include='number')
df_cat=df_House_Price1.select_dtypes(include='object')
```

```python
# Use Dummy Encoding
df_encoded=pd.get_dummies(df_cat)
# Reset the index
df_encoded=df_encoded.reset_index()
```

```python
# import the standard scaler
from sklearn.preprocessing import StandardScaler

# initiate the standard scaler
ss=StandardScaler()
```

```python
# scaled value present in the dataframe
scaled=ss.fit_transform(df_num)
```

```python
# Numerate the data into the dataframe
scaled_df=pd.DataFrame(scaled,columns=df_num.columns)
```

```python
# Reset the index
scaled_df=scaled_df.reset_index()
```

```python
# Concate encoded variable and scaled variable
df_independent=pd.concat([df_encoded,scaled_df],axis=1)
```

```python
# import the train test split
from sklearn.model_selection import train_test_split
```

```python
# split the data
x_train,x_test,y_train,y_test=train_test_split(df_independent,
                                               df_target,
                                               test_size=0.2)
```

**Decision Tree Classifier**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report

# Create a Decision Tree Classifier
tree_model=DecisionTreeClassifier()

# Fit the model to the training data
tree_model.fit(x_train,y_train)

# Predict on the testing data
predictions= tree_model.predict(x_test)

# Calculate classification report
class_report=classification_report(y_test,predictions)
print(class_report)
```
